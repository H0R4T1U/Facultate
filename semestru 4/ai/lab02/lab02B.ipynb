```python
import pandas as pd
from PIL import Image, ImageFilter
import os
import matplotlib.pyplot as plt
# sa se vizualizez una din imagini
path = ("./data/images")
list = os.listdir(path)
img_path = os.path.join(path, list[2])
image = Image.open(img_path)
image.show()
```

```python
dim = (128,128)
imag_redim = []
for imag in list:
    cale_imag = os.path.join(path, imag)
    image = Image.open(cale_imag)
    image = image.resize(dim)
    imag_redim.append(image)

nrImg = len(imag_redim)
nrCol = 3
nrRand = nrImg // nrCol
tabel = pd.DataFrame(columns=[f"Imagine {i+1}" for i in range(nrCol)])
fig, ax = plt.subplots(nrRand, nrCol, figsize=(8,8))
for i, imagine in enumerate(imag_redim):
    ax[i // nrCol, i % nrCol].imshow(imagine)
plt.show()
```

```python
gray = []

for imag in list:
    cale_imag2 = os.path.join(path, imag)
    imagine = Image.open(cale_imag2)
    imagine = imagine.convert("L")
    gray.append(imagine)

nrImagini = len(gray)
nrCol = 2
nrRand = nrImagini // nrCol
tabel = pd.DataFrame(columns=[f"Imagine {i+1}" for i in range(nrCol)])
fig, ax = plt.subplots(nrRand, nrCol, figsize=(8,8))
for i, imag in enumerate(gray):
    ax[i // nrCol, i % nrCol].imshow(imag, cmap="gray")
plt.show()
```

```python
imagine = Image.open(os.path.join(path, list[7]))
blurIntensity = 50
blured_imag = imagine.filter(ImageFilter.GaussianBlur(blurIntensity))
fig, ax = plt.subplots(1, 2, figsize=(8,8))
ax[0].imshow(imagine)
ax[0].set_title("Imagine originala")
ax[1].imshow(blured_imag)
ax[1].set_title("Imagine blurata")
plt.show()
```

```python
image = Image.open(os.path.join(path, list[7]))
edges = image.filter(ImageFilter.FIND_EDGES)
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Afieaz imaginea original
axs[0].imshow(image)
axs[0].set_title('Before')
axs[0].axis('off')

# Afieaz imaginea cu muchiile detectate
axs[1].imshow(edges)
axs[1].set_title('After')
axs[1].axis('off')

plt.show()
```
